# iGEM Human Practice Data Analysis - Box Plots

```python
from __future__ import division, print_function # Makes division and printing work like python 3 (we're using 2)
import os
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from scipy.spatial import distance
from tabulate import tabulate
import csv
%matplotlib inline
```

```python
data_2013 = os.path.join(os.getcwd(), 'datasets', '2013N.csv')
data2013 = pd.read_csv(data_2013)
#data2013.head(3)
```

```python
data_2014 = os.path.join(os.getcwd(), 'datasets', '2014N.csv')
data2014 = pd.read_csv(data_2014)
#data2014.head(3)
```

```python
data_2015 = os.path.join(os.getcwd(), 'datasets', '2015N.csv')
data2015 = pd.read_csv(data_2015)
#data2015.head(3)
```

```python
coor = os.path.join(os.getcwd(), 'datasets', 'coordinates.csv')
coordinates = pd.read_csv(coor)
#coordinates.head(3)
```

```python
#coordinates[0:1][['x', 'y', 'z']].values
```

```python
distance_matrix = distance.cdist(coordinates[0:46][['x', 'y', 'z']].values, coordinates[0:46][['x', 'y', 'z']].values, 'euclidean')
```

```python
p_2013 = data2013.values[:, 4:]
p_2014 = data2014.values[:, 4:]
p_2015 = data2015.values[:, 4:]
```

```python
print(p_2013.shape)
print(p_2014.shape)
print(p_2015.shape)
```

```python
for i in range(104):
    for j in range(44):
        p_2013[i][j+1] /= float(p_2013[i][0])
```

```python
for i in range(113):
    for j in range(44):
        p_2014[i][j+1] /= float(p_2014[i][0])
```

```python
for i in range(153):
    for j in range(44):
        p_2015[i][j+1] /= float(p_2015[i][0])
```

```python
diversity_2013 = []
for i in range(104):
    diversity = 0
    for j in range(1, 44):
        if p_2013[i][j] > 0:
            base = p_2013[i][j]
            for k in range(j+1, 45):
                if p_2013[i][k] > 0:
                    diversity += distance_matrix[j][k] * p_2013[i][j] * p_2013[i][k]
    diversity_2013.append(diversity)

#print(len(diversity_2013))
```

```python
sns.set(style="whitegrid")

f, ax = plt.subplots(figsize=(12, 12))

# Plot the orbital period with horizontal boxes
sns.boxplot(x=data2013.Medal, y=diversity_2013,whis=np.inf, color="powderblue")

# Add in points to show each observation
sns.swarmplot(x=data2013.Medal, y=diversity_2013,size=4, color=".3", linewidth=0)
sns.despine(trim=True, left=True)

plt.xticks(range(4),["Gold", "Silver", "Bronze","None"])
plt.title("Teams of 2013", size=20)
plt.ylabel("Diversity", size=20)
plt.xlabel("Medal", size=20)
plt.tick_params(labelsize=18)
plt.savefig("2013 diversity box plot.png", dpi=300)
plt.show()
```

```python
diversity_2014 = []
for i in range(113):
    diversity = 0
    for j in range(1, 44):
        if p_2014[i][j] > 0:
            base = p_2014[i][j]
            for k in range(j+1, 45):
                if p_2014[i][k] > 0:
                    diversity += distance_matrix[j][k] * p_2014[i][j] * p_2014[i][k]
    diversity_2014.append(diversity)

#print(diversity_2014)
```

```python
sns.set(style="whitegrid")

f, ax = plt.subplots(figsize=(40, 10))

# Plot the orbital period with horizontal boxes
sns.boxplot(x=data2014.Track, y=diversity_2014,whis=np.inf, palette="Set3", width=0.3)

# Add in points to show each observation
sns.swarmplot(x=data2014.Track, y=diversity_2014,size=4, color=".3", linewidth=0)
sns.despine(trim=True, left=True)

# plt.xticks(range(4),["Gold", "Silver", "Bronze","None"])
plt.title("Teams of 2014", size=20)
plt.ylabel("Diversity", size=20)
plt.xlabel("Track", size=20)
plt.tick_params(labelsize=18)
#plt.savefig("2014 diversity diversity box plot.png", dpi=300)
#plt.xticks([])
plt.legend(loc='upper left')
plt.tick_params(axis='x', length=0.1, width=0.1)
plt.show()
```
